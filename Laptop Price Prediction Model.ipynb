```python
def parse_memory(memory_string):
    """
    Parse the memory string from the DataFrame to extract HDD and SSD values.
    This version fixes the issue with converting 'TB' to 'GB' and handles float sizes.

    Parameters:
    - memory_string: a string representing the memory, e.g., "128GB SSD + 1TB HDD"

    Returns:
    - A tuple with two elements, the first is the SSD value, the second is the HDD value.
    """
    # Initialize HDD and SSD values
    hdd_value = 0
    ssd_value = 0

    # Split the memory string by '+' in case there is more than one storage type
    memory_parts = memory_string.split('+')

    # Loop through each part and extract the HDD and SSD values
    for part in memory_parts:
        # Split each part on space to separate size and type
        size_type = part.strip().split(' ')
        
        # Check if there are two elements (size and type)
        if len(size_type) == 2:
            size, storage_type = size_type
            
            # Convert TB to GB if necessary
            if 'TB' in size:
                size = float(size.replace('TB', '')) * 1000  # Convert TB to GB
            else:
                size = float(size.replace('GB', ''))
            
            # Depending on the storage type, add the size to the corresponding value
            if 'HDD' in storage_type:
                hdd_value += size
            elif 'SSD' in storage_type or 'Flash Storage' in storage_type:
                # Assuming Flash Storage to be equivalent to SSD as per common practice
                ssd_value += size
            # Additional storage types like 'Hybrid' can be added here if needed

    return int(ssd_value), int(hdd_value)
```

```python
df[['SSD', 'HDD']] = df['Memory'].apply(lambda x: pd.Series(parse_memory(x)))
```

# Training and Testing

```python
X = df.drop(columns=['Price'])
y = np.log(df['Price'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse=False, drop='first'), 
     [0,1,7,10,11])], remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([('step1', step1),
                ('step2', step2)
                ])

pipe.fit(X_train, y_train)

y_pred = pipe.predict(X_test)

```

```python
from sklearn import metrics
```

```python
print('MAE: ', metrics.mean_absolute_error(y_test, y_pred))
print('MSE: ', metrics.mean_squared_error(y_test, y_pred))
print('RMSE: ', np.sqrt(metrics.mean_absolute_error(y_test, y_pred)))
print('r2 Score: ', metrics.r2_score(y_test, y_pred))
```
